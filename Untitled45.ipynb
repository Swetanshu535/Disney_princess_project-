```python
import pandas as pd #this is done to gain the dataframe of the provided data 
import numpy as np #this is done to gain the numeircla calculations to calculate the important data
import matplotlib.pyplot as plt #this is done to graphically represent the data 
import seaborn as sns #this is done to work on heatmaps and other map structure 
```

```python
#time to read the file 
df=pd.read_csv("C:\\Users\\Swetanshu\\Desktop\\Disney_princess.csv")
print(df)
```

```python
#view the data based on the colomn and rows 
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
#to display the bar plot
sns.countplot(x="FirstMovieTitle",data=df) #setting the x value as FirstMovieTitle and y is countplot
plt.xticks(rotation=45) #this is set to make sure that the x level rotates in a proper way
plt.title("number of princess by film") # this is the title of the graph 
plt.tight_layout() # this is done to make sure that the graph do not overlap to each other
plt.show() #this is written  to show the bar graph
```

```python
#pie chart formation 
df["HasDuet"].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Duet times")
plt.yaxis("")
plt.show()
```
